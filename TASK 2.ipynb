# Task 2: Prediction Using UnSupervised ML

To predict the optimum number of clusters from the 'iris' dataset.

## STEP 1: Importing the data

```python
# Importing the required libraries

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets
from sklearn.cluster import KMeans
```

```python
# Loading the iris data

dataset = datasets.load_iris()
dataset = pd.DataFrame(dataset.data, columns = dataset.feature_names)
```

```python
# Viewing the data

dataset.head()
```

## STEP 2 : Finding the optimum number of clusters


## what is elbow method?

The variations in between no. of clusters within a certain range.

```python
# Calculating the within-cluster sum of square.

x = dataset.iloc[:, [0,1,2,3]].values
within_cluster_sum_of_square = []
clusters_range = range(1,11)
for k in clusters_range:
    km = KMeans(n_clusters = k)
    km = km.fit(x)
    within_cluster_sum_of_square.append(km.inertia_)
```

```python
# Plot the within_cluster sum of square vs clusters range

plt.plot(clusters_range,within_cluster_sum_of_square,'go-', color = 'blue')
plt.title('The Elbow Method')
plt.xlabel('no. of clusters')
plt.ylabel('within-cluster sum of square')
plt.grid()
plt.show()
```

## Step 3 : Applying k-means clustering on the data

```python
from sklearn.cluster import KMeans

model = KMeans(n_clusters = 3,init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
predictions = model.fit_predict(x)
```

## Step 4: Visualising the clusters 

```python
# Visualising the clusters - on the first two columns

plt.scatter(x[predictions ==0,0],x[predictions ==0,1],s=25,c='red',label='Iris-setosa')
plt.scatter(x[predictions ==1,0],x[predictions ==1,1],s=25,c='blue',label='Iris-versicolor')
plt.scatter(x[predictions ==2,0],x[predictions ==2,1],s=25,c='green',label='Iris-virginica' )           

#Plotting the cluster centers

plt.scatter(model.cluster_centers_[:,0],model.cluster_centers_[:,1],s=200,c='black',marker='*', label ='centroids')
plt.legend()
plt.show()
```
